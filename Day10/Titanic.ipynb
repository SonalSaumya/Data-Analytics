```python
#using matplotlib
plt.figure(figsize=(10,6))
plt.hist(df['Age'].dropna(),bins=30, color='blue', alpha=0.8)
plt.title("Distribution of Passenger Ages: ")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(axis='y', alpha= 0.75)
plt.show()
```

```python
#using Seaborn

sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.set(style='darkgrid')
ax= sns.countplot(x='Survived', data= df, palette= 'pastel')
plt.title('Survival Counts')
plt.xlabel('Survived (0= No, 1= Yes)')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0,1], labels= ['No', 'Yes'])
plt.yticks(rotation=45)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
print(survival_counts)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x= 'Fare',data= df, palette= 'pastel' )
plt.title('Distribution of Fare Prices')
plt.xlabel('Fare')
plt.show()
```

```python
#Categorical Variables
#What is the survival rate by passenger class (Pclass)? Use a bar plot.
# How does the survival rate differ by gender? Create a count plot to visualize this.
# What is the survival rate across different embarkation points (Embarked)?
# How does the fare differ across passenger classes (Pclass)? Use a box plot.
```

```python
pclass_survival = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=pclass_survival.index, y=pclass_survival.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Count of Passengers by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()

gender_survival = df.groupby('Sex')['Survived'].mean()
print(gender_survival)
```

```python
embarked_survival = df.groupby('Embarked')['Survived'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=embarked_survival.index, y=embarked_survival.values)
plt.title('Survival Rate by Embarkation Point')
plt.xlabel('Embarkation Point (Embarked)')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data= df, x='Pclass', y='Fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Fare')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.show()
```
